```python
# Ans 3 
""" 
ArithmeticError class is the base class for those built-in exceptions that are raised for various arithmetic errors
1. OverflowError
2. ZeroDivisionError
3. FloatingPointError
exmaples
"""

# ZeroDivisionError
try:
    print(10/0)
except ZeroDivisionError as e:
    print(e)  # not logging it cause it is for testing purpose

# OverflowError
try:
    j = 5.0
    for i in range(1, 1000):
        j = j**i
        print(j)
except OverflowError as e:
    print(e)
```

```python
# Ans 4 
""" 
The LookupError exception in Python forms the base class for all exceptions that are raised when an index or a key is not found for a sequence or dictionary 
"""
# Index Error 
x = [1, 2, 3, 4]
try:
    print(x[10])
except LookupError as e:
    print(f"{e}, {e.__class__}")


# KeyError
user_info = {'name': 'Shanu tyagi','age': 17,'language': 'python'}
user_input = input('What do you want to learn about user : ')
try:
    print(f'{user_input} is {user_info[user_input]}')
except LookupError as e:
    print(f'{e}, {e.__class__}')
```
