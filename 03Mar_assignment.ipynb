```python
# Ans 1 

"""
Seaborn is a popular data visualization library in Python that is built on top of Matplotlib. Here are five common plots that can be created using Seaborn along with their uses:

Scatter plot: This plot is used to visualize the relationship between two continuous variables. It displays points on a two-dimensional plane where each point represents an observation in the dataset. The position of each point is determined by the values of the two variables.

Line plot: This plot is used to visualize the trend or relationship between two variables over time. It displays a line connecting the mean or median values of the variables at each time point.

Histogram: This plot is used to visualize the distribution of a single variable. It displays the frequency of observations falling within certain intervals or bins along the x-axis.

Heatmap: This plot is used to visualize the relationship between two categorical variables. It displays a grid of squares where the color of each square represents the frequency or value of a third variable.

Bar plot: This plot is used to compare the values of a categorical variable across different groups. It displays the mean or median value of the variable for each group as a bar on the y-axis.

These are just a few examples of the many types of plots that can be created using Seaborn. Other common plot types include box plots, violin plots, and regression plots. Seaborn is known for its ease of use and ability to create visually appealing plots quickly and efficiently.
"""
```

```python
# Ans 2 
import seaborn as sns 

fmri = sns.load_dataset('fmri')
sns.lineplot(x= fmri.timepoint, y=fmri.signal)
```

```python
# Ans 3
import matplotlib.pyplot as plt
titanic_df = sns.load_dataset('titanic')
fig, axes = plt.subplots(ncols=2, figsize=(10, 5))

sns.boxplot(x='pclass', y='age', data=titanic_df, ax=axes[0])
sns.boxplot(x='pclass', y='fare', data=titanic_df, ax=axes[1])
plt.show()
```

```python
# Ans 4 

diamond = sns.load_dataset('diamonds')
sns.histplot(data=diamond, x='price', hue=diamond.cut)
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
sns.pairplot(data=iris, hue='species')
```
