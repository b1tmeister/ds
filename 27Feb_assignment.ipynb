```python
# Ans 2 
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.title("This is Title of Question 2")
plt.xlabel("This is X Label")
plt.ylabel("This is Y Label")
plt.show()
```

```python
# Ans 3 

""" 
The subplot() function in Matplotlib is used to create multiple plots (subplots) in a single figure
"""

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

plt.subplot(1,2,1)
plt.plot(x,y)


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

plt.subplot(1,2,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])

plt.subplot(1,2,1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(1,2,1)
plt.plot(x,y)
```

```python
# Ans 4 

""" 
A bar plot is a type of plot that displays categorical data with rectangular bars, where the height of each bar represents the value of the category being plotted. Bar plots are commonly used to compare the values of different categories or to show how the values of a single category vary over time or across different groups
"""
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company, profit)
plt.show()

plt.barh(company, profit)
plt.show()
```
