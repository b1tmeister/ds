```python
# Ans 1 
import seaborn as sns
import plotly.express as px

titanic = sns.load_dataset("titanic")
fig = px.scatter(x=titanic.age, y=titanic.fare)
fig.show()
```

```python
# Ans 2

import plotly.express as px
tips = px.data.tips() 
fig = px.box(tips, x="day", y="total_bill")
fig.show()
```

```python
# Ans 3 

fig = px.histogram(tips, x='sex',y='total_bill', pattern_shape='smoker', color='day' )
fig.show()
```

```python
# Ans 4 

iris = px.data.iris()
fig = px.scatter_matrix(iris, dimensions=['sepal_length','sepal_width','petal_length', 'petal_width'], color='species')
fig.show()
```

```python
# Ans 5 

""" 
Distplot stands for distribution plot is a function in the seaborn package that is used to visualize the distribution of a univariate set of observations. It combines a histogram with a kernel density estimate plot to provide a smooth estimate of the distribution.
"""


import seaborn as sns
import plotly.graph_objects as go


tips = sns.load_dataset("tips")
fig = sns.displot(tips, x="total_bill", kind="kde")

x = fig.axes[0][0].get_lines()[0].get_xdata()
y = fig.axes[0][0].get_lines()[0].get_ydata()

fig2 = go.Figure(data=go.Scatter(x=x, y=y, mode="lines"))
fig2.show()
```
